```python
calculate_distance_matrix = pd.read_csv("C:/Users/pranay/Downloads/dataset-3.csv")
```

```python
calculate_distance_matrix
```

```python
toll_matrix = pd.DataFrame(0, index=calculate_distance_matrix['id_start'].unique(), columns=calculate_distance_matrix['id_start'].unique())
```

```python
for i, row in calculate_distance_matrix.iterrows():
    toll_matrix.at[row['id_end'], row['id_start']] = row['distance']
    toll_matrix.at[row['id_start'], row['id_end']] = row['distance']
```

```python
 toll_matrix.head()
```

```python
matrix_size=43
for i in range(matrix_size):
    for j in range(matrix_size):
        if i != j and toll_matrix.iloc[i, j] == 0:
            toll_matrix.iloc[i, j] = toll_matrix.iloc[i-1, j] + toll_matrix.iloc[i, i-1]
```

```python
matrix_size=43
for i in range(matrix_size):
    for j in range(matrix_size):
        if i > j :
            toll_matrix.iloc[j,i] = toll_matrix.iloc[i,j] 
```
